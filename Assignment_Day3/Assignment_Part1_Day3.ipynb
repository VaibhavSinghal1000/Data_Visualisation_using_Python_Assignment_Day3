```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic=titanic.dropna()
```

```python
titanic.info()
```

```python
titanic.head()
```

```python
x=titanic['fare']
plt.plot(x)
plt.xlabel('Survived')
```

```python
x=titanic['survived']
y=titanic['pclass']
plt.plot(x,y)
plt.xlabel('Survived')
plt.ylabel('Pclass')
```

```python
plt.scatter(x='age',y='fare',data=titanic,alpha=0.7)
```

```python
plt.bar(titanic['age'],titanic['fare'])
```

```python
plt.hist(titanic['fare'],histtype='step',color='r')
```

```python
plt.boxplot(titanic['age'])
```

```python
plt.boxplot(titanic['fare'])
```

```python
plt.figure(figsize=(22,12),dpi=500)

plt.subplot(2,3,1)
plt.plot(titanic['embark_town'],titanic['alive'])
plt.xlabel('Embark Town')
plt.ylabel('Alive')

plt.subplot(2,3,2)
plt.plot(titanic['embark_town'],titanic['embarked'])
plt.xlabel('Embark Town')
plt.ylabel('Embarked')

plt.subplot(2,3,3)
plt.scatter(titanic['age'],titanic['embark_town'])
plt.xlabel('Age')
plt.ylabel('Embark Town')

plt.subplot(2,3,4)
plt.hist(titanic['embark_town'])
plt.xlabel('Embark Town')

plt.subplot(2,3,5)
plt.bar(titanic['embark_town'],titanic['age'])
plt.xlabel('Embark Town')
plt.ylabel('Age')

plt.subplot(2,3,6)
plt.boxplot(titanic['age'])
```
